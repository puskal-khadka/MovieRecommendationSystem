```python
# @author: https://github.com/puskal-khadka

import pandas as pd

df_credits=pd.read_csv("dataset/tmdb_5000_credits.csv")
df_movies=pd.read_csv("dataset/tmdb_5000_movies.csv")

#merge two dataframe into single 
df=df_movies.merge(df_credits,on="title")

df.head(1)
```

```python
#for this engine we are taking  overview, genres, keyboard, cast consideration for recommendation
df=df[["id","title","genres","keywords","cast","overview"]]
df.head()
```

```python
df["overview"]=df["overview"].astype(str).str.split()
df["overview"]
```

```python
def convertData():
    considered_feature=["genres","keywords","cast"]
    for c in considered_feature:
        df[c]=df[c].apply(preProcess)
                
def preProcess(obj):
    import ast
    tempList=[]
    count=0
    for i in ast.literal_eval(obj):
        if count>=4:  #maximum 4 words for each column
            break
        tempList.append(i['name'])
        count+=1   
    return tempList


convertData()

df   #visualize final converted data frame  
        
```

```python
# merge considered attribute
df['filteredWord']=df['overview'] + df['genres']+df['keywords']+df['cast']
df['filteredWord']=df.filteredWord.apply(lambda i: " ".join(i) )
df.head()
```

```python
#final processing for filteredWord

from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import re

# Function for removing NonAscii characters
def _removeNonAscii(s):
    return "".join(i for i in s if  ord(i)<128)
# Function for converting into lower case
def make_lower_case(text):
    return text.lower()
# Function for removing stop words
def remove_stop_words(text):
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text
# Function for removing punctuation
def remove_punctuation(text):
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    text = " ".join(text)
    return text
#Function for removing the html tags
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

# Applying all the functions in description and storing as a filteredWord

df['filteredWord'] = df['filteredWord'].apply(_removeNonAscii)
df['filteredWord'] = df.filteredWord.apply(func = make_lower_case)
df['filteredWord'] = df.filteredWord.apply(func = remove_stop_words)
df['filteredWord'] = df.filteredWord.apply(func=remove_punctuation)
df['filteredWord'] = df.filteredWord.apply(func=remove_html)

df['filteredWord']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(ngram_range=(1,3), stop_words='english')
tf_matrix=tfidf.fit_transform(df['filteredWord'])
tf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity=cosine_similarity(tf_matrix,tf_matrix)

cosine_similarity[0] #debug, seeing first movie score
```

```python
indices=pd.Series(df.index, index=df['title']).drop_duplicates()

indices
```

```python
def movieRecommend(title):
    index=indices[title]
    similarity_score=cosine_similarity[index]
    enumerated_sim_score= list(enumerate(similarity_score))
    recommended_movieList=sorted(enumerated_sim_score, key=lambda x:x[1],reverse=True)[1:6] # 5 movie
  
    movie_indices=[ i[0] for i in recommended_movieList ]
    recommend=df.iloc[movie_indices].title
    return recommend 
    
```

```python
movieRecommend('The Avengers')
```
